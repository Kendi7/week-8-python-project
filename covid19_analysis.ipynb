```python
sns.scatterplot(
    data = covid_data.groupby('WHO Region').agg({
        'Confirmed': 'sum',
        'Deaths': 'sum'
    }).reset_index(),
    x='Deaths',
    y='Confirmed',
    hue = 'WHO Region'
)
plt.title('Confirmed vs Deaths toll')
```

```python
sns.lineplot(data=covid_data, x='1 week change', y='Deaths')
plt.title('Deaths agains  weeks change')
```

```python
sns.barplot( 
    data = covid_data.groupby('WHO Region')['Deaths'].sum().reset_index(),
 x = 'Deaths',
 y = 'WHO Region'
)

plt.title('Representation of Regions against Deaths Toll')
```

```python
# Sort and select top 5 countries
top5 = covid_data.sort_values(by='Confirmed', ascending=False).head(5)

# Plot bar graph
sns.barplot(
    data=top5,
    x='Country/Region',
    y='Confirmed',
    palette='mako'  # Optional color theme
)

plt.title('Top 5 Countries by Confirmed COVID-19 Cases')
plt.xlabel('Country')
plt.ylabel('Confirmed Cases')
plt.xticks(rotation=45)
```

```python
# Select only numerical columns
numeric_data= covid_data.select_dtypes(include=['number'])

# Correlation heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of COVID-19 Metrics")
plt.show()
```

```python
#Calculate the death rate: total_deaths / total_cases.
total_death_case = covid_data['Deaths'].sum()
total_confirmed_case = covid_data['Confirmed'].sum()
death_rate = total_death_case /total_confirmed_case
```

```python
#Analysing vaccination data  uploaded
# Get latest data
latest = covid_data_vaccination.sort_values('date').drop_duplicates('country', keep='last')

# Top 10 countries by % vaccinated
top_vax = latest.sort_values('daily_vaccinations_per_million', ascending=False).head(10)

# Bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data=top_vax, x='country', y='daily_vaccinations_per_million')
plt.title('Top 10 Countries last days of daily_vaccinations_per_million')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
